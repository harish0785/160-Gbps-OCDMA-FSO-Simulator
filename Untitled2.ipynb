```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import (accuracy_score, confusion_matrix, ConfusionMatrixDisplay,
                             roc_curve, auc, mean_squared_error, r2_score)

# 1. Load the dataset
# If running in Colab, ensure you have uploaded 'Manufacturing_dataset.csv'
file_path = 'Manufacturing_dataset.csv'
df = pd.read_csv(file_path)

# Drop Timestamp for modeling
data = df.drop(columns=['Timestamp'])

# ---------------------------------------------------------
# PART 1: CLASSIFICATION (Target: Optimal Conditions)
# ---------------------------------------------------------
X_clf = data.drop(columns=['Optimal Conditions'])
y_clf = data['Optimal Conditions']

X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_clf, y_clf, test_size=0.2, random_state=42)

clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train_c, y_train_c)
y_pred_c = clf.predict(X_test_c)
y_prob_c = clf.predict_proba(X_test_c)[:, 1]

# --- 1. Accuracy Graph (Learning Curve) ---
train_sizes, train_scores, test_scores = learning_curve(clf, X_clf, y_clf, cv=5, scoring='accuracy', n_jobs=-1)
train_mean = np.mean(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)

plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_mean, label='Training Accuracy')
plt.plot(train_sizes, test_mean, label='Cross-Validation Accuracy')
plt.title('Accuracy Graph (Learning Curve)')
plt.xlabel('Training Data Size')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

# --- 2. Confusion Matrix ---
cm = confusion_matrix(y_test_c, y_pred_c)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# --- 3. ROC Curve ---
fpr, tpr, _ = roc_curve(y_test_c, y_prob_c)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()

print(f"Classification Accuracy: {accuracy_score(y_test_c, y_pred_c):.4f}")

# ---------------------------------------------------------
# PART 2: REGRESSION (Target: Production Quality Score)
# ---------------------------------------------------------
X_reg = data.drop(columns=['Production Quality Score'])
y_reg = data['Production Quality Score']

X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

reg = RandomForestRegressor(n_estimators=100, random_state=42)
reg.fit(X_train_r, y_train_r)
y_pred_r = reg.predict(X_test_r)

# --- 4. R2 Graph (Actual vs Predicted) ---
plt.figure(figsize=(8, 6))
plt.scatter(y_test_r, y_pred_r, alpha=0.5, color='green')
plt.plot([y_test_r.min(), y_test_r.max()], [y_test_r.min(), y_test_r.max()], 'r--', lw=2)
plt.title('R2 Graph: Actual vs Predicted')
plt.xlabel('Actual Quality Score')
plt.ylabel('Predicted Quality Score')
plt.show()

# --- 5 & 6. MSE and RMSE ---
mse = mean_squared_error(y_test_r, y_pred_r)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_r, y_pred_r)

print("-" * 30)
print(f"Mean Square Error (MSE): {mse:.4f}")
print(f"Root Mean Square Error (RMSE): {rmse:.4f}")
print(f"R2 Score: {r2:.4f}")
```
